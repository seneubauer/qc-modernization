```python
def interpret_receiver_number(x):
    if type(x) is str:
        if "|" in x:
            return x.split("|")
        else:
            return [x]
    else:
        return [None]
```

```python
def interpret_supplier(x, reference_dictionary):
    for i in reference_dictionary:
        if i["key"] == x:
            return { "id": i["index"], "name": i["value"] }
    return { "id": 0, "name": "n/a" }
```
